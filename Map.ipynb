```python
from google.colab import drive
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
austin_df = pd.read_csv("/content/drive/MyDrive/Data Visualization /listings.csv")
austin_df.head()
```

##Part I

```python
# Drop rows with missing latitude, longitude, or price
austin_df = austin_df.dropna(subset=['latitude', 'longitude', 'price'])

# Convert price column to numeric
austin_df['price'] = austin_df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Plot the bubble map
fig = px.scatter_mapbox(austin_df,
                        title="Accommodation Prices in Austin, Texas",
                        lat='latitude',
                        lon='longitude',
                        size='price',
                        color='price',
                        color_continuous_scale='Blugrn',
                        zoom=10,
                        mapbox_style='carto-positron',
                        size_max=80,
                        hover_data=['price', 'host_name', 'name'],
                        height=800,
                        labels={"price": "Price (USD)"})

fig.update_layout(margin=dict(l=0, r=0, t=55, b=0))

fig.show()
```

##Part II

```python
# Define major cities along the Silk Route
silk_route_cities = {
    "Xi'an": (108.9398, 34.3416),
    "Samarkand": (66.9719, 39.6270),
    "Bukhara": (64.4280, 39.7680),
    "Tashkent": (69.2669, 41.2995),
    "Tbilisi": (44.7866, 41.7151),
    "Ankara": (32.8597, 39.9334),
    "Istanbul": (28.9784, 41.0082),
    "Damascus": (36.2020, 33.5138),
    "Baghdad": (44.3615, 33.3152),
    "Cairo": (31.2357, 30.0444),
    "Kashgar": (75.9797, 39.4704),
    "Kandahar": (65.7381, 31.5497),
}
```

```python
city_names, lats, lons = zip(*[(city, lat, lon) for city, (lon, lat) in silk_route_cities.items()])

city_trace = go.Scattergeo(
    lon=lons,
    lat=lats,
    text=city_names,
    mode="markers",
    marker=dict(size=10, color='green', symbol='circle'),
    name="Cities"
)

silk_route_trace = go.Scattergeo(
    lon=[lon for lon, _ in silk_route_cities.values()],
    lat=[lat for _, lat in silk_route_cities.values()],
    mode="lines",
    line=dict(width=2, color='orange', dash='dot'),
    name="Silk Route"
)
```

```python
# Define layout
layout = go.Layout(
    title="Simplified Silk Route with Major Cities",
    geo=dict(
        resolution=110,
        showland=True,
        showocean=True,
        showcountries=True,
        countrycolor='rgb(204, 204, 204)',
        coastlinecolor='rgb(0, 0, 0)',
        oceancolor='rgb(0, 255, 255)',
        projection_type='natural earth',
        landcolor='rgb(240, 240, 240)'
    )
)

fig = go.Figure(data=[city_trace, silk_route_trace], layout=layout)

fig.show()
```
